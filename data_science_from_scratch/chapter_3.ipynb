```python
import random
import numpy as np
from matplotlib import pyplot as plt

Xs = np.random.randn(5)*100
ys = [10,20,30,40,50]

c = 'yellow'
plt.plot(Xs, ys, color='green', marker='o', linestyle='solid')
plt.title('Random data', color=c)
plt.ylabel('Y Axis Values', color=c)
plt.xlabel('X Axis Values', color=c)
plt.show()
```

## Bar Charts

```python
Xs = ['Cat 1', 'Cat 2', 'Cat 3']
ys = [10,20,30]
c = 'cyan'
plt.bar(range(len(Xs)), ys)
plt.yticks(color=c)
plt.xticks(range(len(Xs)), Xs, color=c)
plt.xlabel('Categories', color=c)
plt.ylabel('Arbitrary values', color=c)
```

### histogram

```python
from collections import Counter
c = 'cyan'
grades = [90, 77, 99, 0, 89, 94, 100, 75, 88, 100, 0, 0]
histogram = Counter((min(grade // 10 * 10, 100) for grade in grades))
plt.bar([x+5 for x in histogram.keys()], histogram.values(), width=10, edgecolor='black', color=c)
plt.axis([-5, 115, 0, 5])
plt.xticks([i * 10 for i in range(11)])
plt.ylabel('No. of students', color=c)
plt.xlabel('Percentile', color=c)
```

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias = [x for x in reversed(variance)]
total_error = [x+y for x, y in (zip(variance, bias))]
Xs = [i for i in range(len(variance))]

plt.plot(Xs, variance, 'g-', label='variance')
plt.plot(Xs, bias, 'r-', label='bias^2')
plt.plot(Xs, total_error, 'b:', label="total error")
plt.legend(loc=9)
plt.title('Bias vs Variance', size=10, color='yellow')
plt.xticks(color='yellow')
plt.yticks(color='yellow')
```

```python
friends = [40, 50, 65, 55, 35, 122, 99, 125, 36]
minutes = [123, 90, 100, 99, 45, 200, 100, 200, 40]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)

for friend_count, minute_count, label in zip(friends, minutes, labels):
    plt.annotate(label, xy=(friend_count, minute_count), xytext=(5, -5), textcoords='offset points')

plt.ylabel('Minutes per month', color='cyan')
plt.xlabel('Friend count', color='cyan')
```
